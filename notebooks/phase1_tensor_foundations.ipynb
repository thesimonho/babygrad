# Phase 1: Tensor Foundations

Small examples for the completed tensor foundation features.

```python
from babygrad.tensor import Tensor
from babygrad.types import NodeKind
```

## Tensor storage, shape, and indexing

A tensor stores values flat while its shape decides how indexing interprets those values.

```python
scores = Tensor([1, 2, 3, 4, 5, 6], shape=(2, 3), kind=NodeKind.INPUT)

{
    "tensor": scores,
    "shape": scores.shape,
    "rank": scores.ndim,
    "items": scores.numel,
    "first_row": scores[0],
    "bottom_right": scores[-1, -1],
}
```

## Element-wise operations and broadcasting

Element-wise operations preserve the aligned shape. Bias-shaped tensors can broadcast across matrix rows.

```python
features = Tensor([1, 2, 3, 4, 5, 6], shape=(2, 3), kind=NodeKind.INPUT)
row_bias = Tensor([10, 20, 30], shape=(1, 3), kind=NodeKind.INPUT)

shifted = features + row_bias
scaled = shifted * Tensor([2], shape=(1, 1), kind=NodeKind.INPUT)
rooted = Tensor([1, 4, 9, 16], shape=(2, 2), kind=NodeKind.INPUT).sqrt()
logged = Tensor([1, 10], shape=(1, 2), kind=NodeKind.INPUT).log()

{
    "shifted": shifted,
    "scaled": scaled,
    "sqrt": rooted,
    "log": logged,
}
```

## Reductions

Reductions keep rank by turning the reduced axis into size 1.

```python
batch = Tensor([1, 2, 3, 4, 5, 6], shape=(2, 3), kind=NodeKind.INPUT)

{
    "column_totals": batch.sum(axis=0),
    "row_totals": batch.sum(axis=1),
    "last_axis_totals": batch.sum(axis=-1),
    "overall_mean": batch.mean(),
}
```

## Shape manipulation

Reshape and flatten keep the values in row-major order. Transpose rearranges the 2D layout.

```python
matrix = Tensor([1, 2, 3, 4, 5, 6], shape=(2, 3), kind=NodeKind.INPUT)

{
    "original": matrix,
    "reshaped": matrix.reshape((3, 2)),
    "flattened": matrix.flatten(),
    "transposed": matrix.transpose(),
}
```

## Matrix multiplication

Matmul combines the left rows with the right columns.

```python
left = Tensor([1, 2, 3, 4, 5, 6], shape=(2, 3), kind=NodeKind.INPUT)
right = Tensor([7, 8, 9, 10, 11, 12], shape=(3, 2), kind=NodeKind.INPUT)

left @ right
```
